```python
# Load library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error
from matplotlib.dates import DateFormatter
from datetime import datetime
```

```python
model_labels = {
    "y_true": "Ture Y",
    "MLP_level": "MLP predictions",
    "LSTM_level": "LSTM predictions",
}

model_colors = {
    "Ture Y": "#000000",              # black
    "MLP predictions": "#999999",     # gray60
    "LSTM predictions": "#B3B3B3"     # gray70
}

model_linetypes = {
    "Ture Y": "solid",
    "MLP predictions": "dotted",
    "LSTM predictions": "dashed"
}

# === simulation_1.pdf ===
df1 = pd.read_csv("comparison_A_stationary.csv")
plt.figure(figsize=(7, 5))
for col, label in model_labels.items():
    plt.plot(pd.to_datetime(df1['DATE']), df1[col],
             label=label,
             color=model_colors[label],
             linestyle=model_linetypes[label])
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.ylim(50, 150)  
plt.savefig("simulation_1.pdf", bbox_inches='tight', dpi=500)
plt.show()

# === simulation_2.pdf ===
df2 = pd.read_csv("comparison_B_stationary.csv")
plt.figure(figsize=(7, 5))
for col, label in model_labels.items():
    plt.plot(pd.to_datetime(df2['DATE']), df2[col],
             label=label,
             color=model_colors[label],
             linestyle=model_linetypes[label])
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.ylim(50, 150)  
plt.savefig("simulation_2.pdf", bbox_inches='tight', dpi=500)
plt.show()

# === simulation_3.pdf ===
df3 = pd.read_csv("comparison_A_nonstationary.csv")
plt.figure(figsize=(7, 5))
for col, label in model_labels.items():
    plt.plot(pd.to_datetime(df3['DATE']), df3[col],
             label=label,
             color=model_colors[label],
             linestyle=model_linetypes[label])
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.ylim(50, 750)  
plt.savefig("simulation_3.pdf", bbox_inches='tight', dpi=500)
plt.show()

# === simulation_4.pdf ===
df4 = pd.read_csv("comparison_B_nonstationary.csv")
plt.figure(figsize=(7, 5))
for col, label in model_labels.items():
    plt.plot(pd.to_datetime(df4['DATE']), df4[col],
             label=label,
             color=model_colors[label],
             linestyle=model_linetypes[label])
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.ylim(50, 150)  
plt.savefig("simulation_4.pdf", bbox_inches='tight', dpi=500)
plt.show()
```

```python
df_list = [
    ("A_stationary", pd.read_csv("comparison_A_stationary.csv")),
    ("B_stationary", pd.read_csv("comparison_B_stationary.csv")),
    ("A_nonstationary", pd.read_csv("comparison_A_nonstationary.csv")),
    ("B_nonstationary", pd.read_csv("comparison_B_nonstationary.csv")),
]

models = ['MLP', 'LSTM']
results = []

for df_name, df in df_list:
    for model in models:
        level_col = f"{model}_level"
        diff_col = f"{model}_diff"

        valid_level = df[['y_true', level_col]].dropna()
        y_true = valid_level['y_true']
        y_pred = valid_level[level_col]

        mae = mean_absolute_error(y_true, y_pred)
        rmse = mean_squared_error(y_true, y_pred) ** 0.5
        corr = np.corrcoef(y_true, y_pred)[0, 1]

        valid_diff = df[['y_diff', diff_col]].dropna()
        sign_acc = (np.sign(valid_diff['y_diff']) == np.sign(valid_diff[diff_col])).mean()

        results.append({
            "Dataset": df_name,
            "Model": model,
            "MAE": mae,
            "RMSE": rmse,
            "Correlation": corr,
            "Sign Accuracy": sign_acc
        })

results_df = pd.DataFrame(results)
print(results_df.round(4))
```
